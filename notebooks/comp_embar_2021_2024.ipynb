# Análise Comparativa de Embarcações (2021 vs 2024)

O objetivo desta análise é **comparar a quantidade de embarcações por tipo** entre os anos de **2021 e 2024**, para identificar mudanças no perfil da marina ao longo do tempo.

Com isso, será possível:
- Ver quais tipos de embarcação **aumentaram** ou **diminuíram**
- Medir a **diferença total** e a **variação percentual** por categoria
- Destacar tendências gerais de crescimento, queda ou estabilidade no período

```python
# Passo a Passo do projeto

# Passo 1: Abrir a base de dados (Importar a base de dados)
import pandas as pd

table_2021 = pd.read_excel("../data/total_tipo_emb_2021.xlsx")
table_2024 = pd.read_excel("../data/total_tipo_emb_2024.xlsx")

# Passo 2: Visualizar a base de dados
    # entender as informações disponiveis
    # possiveis problemas/erro na base de dados
display(table_2021)
display(table_2024)
```

```python
# Passo 3: Corrigir os problemas da base de dados (tratamento de dados)

# valores em formatos errados
    # nenhum

# informações vazias
table_2021 = table_2021.dropna()
table_2024 = table_2024.dropna()

display(table_2021.info())
display(table_2024.info())
```

```python
# Passo 4: Analise Inicial (Entender cada categoria e as quantidades)

# quantidade - contar quantidade total de embarcações por tipo
res_2021 = (table_2021
            .groupby("TIPO_EMBARCACAO", as_index=False)["QUANTIDADE"]
            .sum()
            .sort_values("QUANTIDADE", ascending=False)
            )

res_2024 = (table_2024
            .groupby("TIPO_EMBARCACAO", as_index=False)["QUANTIDADE"]
            .sum()
            .sort_values("QUANTIDADE", ascending=False)
            )
# verificar somente as 10 primeiras
display(res_2021.head(10))
display(res_2024.head(10))

```

```python
# Passo 4: Tema visual e utilitarios de grafico

import plotly.express as px
import plotly.graph_objects as go

PALETA = ["#0B3C5D", "#328CC1", "#D9B310", "#1D2731", "#B1D4E0"]

def aplicar_tema(fig, titulo, x_titulo="", y_titulo=""):
    fig.update_layout(
        title=dict(text=titulo, x=0.02, xanchor="left", font=dict(size=20)),
        template="plotly_white",
        colorway=PALETA,
        font=dict(family="Segoe UI", size=13, color="#1f2937"),
        xaxis_title=x_titulo,
        yaxis_title=y_titulo,
        plot_bgcolor="white",
        paper_bgcolor="white",
        margin=dict(l=40, r=20, t=70, b=40),
        legend_title_text=""
    )
    fig.update_xaxes(showgrid=False, zeroline=False)
    fig.update_yaxes(showgrid=True, gridcolor="#e5e7eb", zeroline=False)
    return fig

# Ranking Top 10 - 2021
top_2021 = res_2021.head(10).sort_values("QUANTIDADE")
fig = px.bar(
    top_2021,
    x="QUANTIDADE",
    y="TIPO_EMBARCACAO",
    orientation="h",
    text="QUANTIDADE",
    color="QUANTIDADE",
    color_continuous_scale=["#B1D4E0", "#0B3C5D"]
)
fig = aplicar_tema(fig, "Top 10 tipos de embarcacao - 2021", "Quantidade", "Tipo de embarcacao")
fig.update_traces(textposition="outside", hovertemplate="%{y}<br>Quantidade: %{x:,}<extra></extra>")
fig.show()

# Ranking Top 10 - 2024
top_2024 = res_2024.head(10).sort_values("QUANTIDADE")
fig = px.bar(
    top_2024,
    x="QUANTIDADE",
    y="TIPO_EMBARCACAO",
    orientation="h",
    text="QUANTIDADE",
    color="QUANTIDADE",
    color_continuous_scale=["#B1D4E0", "#0B3C5D"]
)
fig = aplicar_tema(fig, "Top 10 tipos de embarcacao - 2024", "Quantidade", "Tipo de embarcacao")
fig.update_traces(textposition="outside", hovertemplate="%{y}<br>Quantidade: %{x:,}<extra></extra>")
fig.show()
```

```python
# Passo 4

# Ajustar escrita diferente entre os anos
table_2021["TIPO_EMBARCACAO"] = table_2021["TIPO_EMBARCACAO"].astype(str).str.strip()
table_2024["TIPO_EMBARCACAO"] = table_2024["TIPO_EMBARCACAO"].astype(str).str.strip()


# Quantos tipos existem em cada ano
display("Tipos em 2021:", table_2021["TIPO_EMBARCACAO"].nunique())
display("Tipos em 2024:", table_2024["TIPO_EMBARCACAO"].nunique())

# Quais tipos são (lista)
tipos_2021 = sorted(table_2021["TIPO_EMBARCACAO"].unique())
tipos_2024 = sorted(table_2024["TIPO_EMBARCACAO"].unique())

display("Lista 2021:")
display(tipos_2021)

display("\nLista 2024:")
display(tipos_2024)

# O que tem em um ano e não teve no outro
so_2021 = sorted(set(tipos_2021) - set(tipos_2024))
so_2024 = sorted(set(tipos_2024) - set(tipos_2021))

display("Só em 2021:", so_2021)
display("Só em 2024:", so_2024)
```

```python
# Passo 5: Analise Detalhada ( Principais diferenças)
import pandas as pd

# (opcional, mas ajuda) padronizar textos
table_2021["TIPO_EMBARCACAO"] = table_2021["TIPO_EMBARCACAO"].astype(str).str.strip()
table_2024["TIPO_EMBARCACAO"] = table_2024["TIPO_EMBARCACAO"].astype(str).str.strip()

# garantir quantidade numérica
table_2021["QUANTIDADE"] = pd.to_numeric(table_2021["QUANTIDADE"], errors="coerce").fillna(0)
table_2024["QUANTIDADE"] = pd.to_numeric(table_2024["QUANTIDADE"], errors="coerce").fillna(0)

# 1) Agrupar por tipo (uma tabela por ano)
qtd_2021 = table_2021.groupby("TIPO_EMBARCACAO", as_index=False)["QUANTIDADE"].sum()
qtd_2024 = table_2024.groupby("TIPO_EMBARCACAO", as_index=False)["QUANTIDADE"].sum()

# 2) Renomear colunas pra não confundir no merge
qtd_2021 = qtd_2021.rename(columns={"QUANTIDADE": "QTD_2021"})
qtd_2024 = qtd_2024.rename(columns={"QUANTIDADE": "QTD_2024"})

# 3) Merge (outer = pega todos os tipos dos dois anos)
comparacao = pd.merge(qtd_2021, qtd_2024, on="TIPO_EMBARCACAO", how="outer")

# 4) Tudo que não existir em um ano vira 0
comparacao[["QTD_2021", "QTD_2024"]] = comparacao[["QTD_2021", "QTD_2024"]].fillna(0).astype(int)

# conferir
display(comparacao.sort_values("QTD_2024", ascending=False).head(20))

```

```python
# Passo 5: Calcular a diferença e a variação

# Diferença absoluta
comparacao["DIF"] = comparacao["QTD_2024"] - comparacao["QTD_2021"]

# Variação percentual (evita divisão por zero)
comparacao["VAR_%"] = ((comparacao["DIF"] / comparacao["QTD_2021"].replace(0, pd.NA)) * 100).round(2)

# Quando 2021 = 0, a variação % não faz sentido -> coloca como 0 (ou pode deixar como NA)
comparacao["VAR_%"] = comparacao["VAR_%"].fillna(0)

# Status
comparacao["STATUS"] = comparacao["DIF"].apply(
    lambda x: "Aumentou" if x > 0 else ("Diminuiu" if x < 0 else "Estável")
)

# Ver as maiores mudanças
display(comparacao.sort_values("DIF", ascending=False).head(10))  # top aumentos
display(comparacao.sort_values("DIF").head(10))                  # top quedas
```

```python
# Parte 5: Novos ou Sumiram em 2024
novos_2024 = comparacao[(comparacao["QTD_2021"] == 0) & (comparacao["QTD_2024"] > 0)]
sumiram_2024 = comparacao[(comparacao["QTD_2021"] > 0) & (comparacao["QTD_2024"] == 0)]

display(novos_2024.sort_values("QTD_2024", ascending=False))
display(sumiram_2024.sort_values("QTD_2021", ascending=False))
```

```python
# Passo 5: Graficos de maiores altas e maiores quedas

# Top 10 aumentos
top_up = comparacao.sort_values("DIF", ascending=False).head(10).sort_values("DIF")

fig = px.bar(
    top_up,
    x="DIF",
    y="TIPO_EMBARCACAO",
    orientation="h",
    title="Top 10 aumentos (2024 - 2021)",
    text="DIF",
    color_discrete_sequence=["#16A34A"]
)
fig = aplicar_tema(fig, "Top 10 aumentos (2024 - 2021)", "Diferenca (QTD_2024 - QTD_2021)", "Tipo de embarcacao")
fig.add_vline(x=0, line_width=1, line_dash="dash", line_color="#6b7280")
fig.update_traces(textposition="outside", hovertemplate="%{y}<br>Diferenca: %{x:,}<extra></extra>")

fig.write_html("../data/comparativo_aumento_2021_2024.html")
fig.show()

# Top 10 quedas
top_down = comparacao.sort_values("DIF").head(10).sort_values("DIF")

fig = px.bar(
    top_down,
    x="DIF",
    y="TIPO_EMBARCACAO",
    orientation="h",
    title="Top 10 quedas (2024 - 2021)",
    text="DIF",
    color_discrete_sequence=["#DC2626"]
)
fig = aplicar_tema(fig, "Top 10 quedas (2024 - 2021)", "Diferenca (QTD_2024 - QTD_2021)", "Tipo de embarcacao")
fig.add_vline(x=0, line_width=1, line_dash="dash", line_color="#6b7280")
fig.update_traces(textposition="outside", hovertemplate="%{y}<br>Diferenca: %{x:,}<extra></extra>")

fig.write_html("../data/comparativo_quedas_2021_2024.html")
fig.show()
```

```python
# Passo 5: Comparativo 2021 vs 2024 (Para os tipos com maior mudanca)
top_changes = (comparacao
               .assign(ABS_DIF=lambda d: d["DIF"].abs())
               .sort_values("ABS_DIF", ascending=False)
               .head(10)
               .sort_values("QTD_2024"))

fig = go.Figure()

for _, r in top_changes.iterrows():
    fig.add_trace(go.Scatter(
        x=[r["QTD_2021"], r["QTD_2024"]],
        y=[r["TIPO_EMBARCACAO"], r["TIPO_EMBARCACAO"]],
        mode="lines",
        line=dict(color="#9ca3af", width=3),
        showlegend=False,
        hoverinfo="skip"
    ))

fig.add_trace(go.Scatter(
    x=top_changes["QTD_2021"],
    y=top_changes["TIPO_EMBARCACAO"],
    mode="markers",
    marker=dict(size=10, color="#2563EB"),
    name="2021",
    hovertemplate="%{y}<br>2021: %{x:,}<extra></extra>"
))

fig.add_trace(go.Scatter(
    x=top_changes["QTD_2024"],
    y=top_changes["TIPO_EMBARCACAO"],
    mode="markers",
    marker=dict(size=10, color="#D97706"),
    name="2024",
    hovertemplate="%{y}<br>2024: %{x:,}<extra></extra>"
))

fig = aplicar_tema(fig, "Top 10 maiores mudancas - comparacao 2021 vs 2024", "Quantidade", "Tipo de embarcacao")
fig.show()
```

```python
# Passo 5: Exportar dados de comparação e exportar o gráfico para HTML
# comparacao deve conter: TIPO_EMBARCACAO, QTD_2021, QTD_2024, DIF, VAR_%, STATUS

# CSV (recomendado para compartilhar)
comparacao.to_csv("../data/comparacao_embarcacoes_2021_2024.csv", index=False, encoding="utf-8-sig")

# Excel (.xlsx)
comparacao.to_excel("../data/comparacao_embarcacoes_2021_2024.xlsx", index=False)
```
